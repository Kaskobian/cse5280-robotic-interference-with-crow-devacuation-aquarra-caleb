# 6. Robot Arm Model (Forward & Inverse Kinematics)

This class defines the robot arm used in the simulation:
### Forward kinematics
Computes the end-effector position from joint angles:
$$
e = f(\phi_1, \phi_2, \phi_3, \phi_4)
$$

### Numerical Jacobian
Approximates:
$$
J = \frac{\partial e}{\partial \phi}
$$
using finite differences.

### Inverse kinematics (IK)
Updates joint angles using:
$$
\Delta \phi = J^+ \cdot (e_{target} - e)
$$
where $J^+$ is the pseudoinverse of the Jacobian.

### Reactive IK
Each timestep performs small updates:
- keeps motion smooth
- avoids instability
- allows real-time control

### End-effector projection
The 3D end-effector position is projected into 2D:
$$
(x, y, z) \rightarrow (x, y)
$$
to act as a dynamic obstacle in the crowd simulation.

This class bridges continuous robot motion and discrete crowd interaction.

```python
# 6. Robot arm model

class RobotArm:
    """
    Robot arm with:
      - forward kinematics
      - Jacobian-based IK
      - persistent vedo meshes updated in-place
      - single-step IK for reactive control in the crowd simulation
    """

    def __init__(self, partLengths, parts, arm_location):
        self.arm_location = np.array(arm_location, dtype=float).reshape(3, 1)

        self.L1, self.L2, self.L3, self.L4 = partLengths
        self.source_parts = parts

        # IK parameters
        self.delta_phi = 0.1
        self.target = np.array([0.0, 100.0, 200.0], dtype=float)

        self.target_lambda = 0.02
        self.convergence = 0.02

        self.meshes = None
        self.initialize_meshes()

    def initialize_meshes(self):
        Base = self.source_parts[0].clone()
        Part1 = self.source_parts[1].clone()
        Part2 = self.source_parts[2].clone()
        Part3 = self.source_parts[3].clone()
        Part4 = self.createCoordinateFrameMesh()
        self.meshes = [Base, Part1, Part2, Part3, Part4]

    def RotationMatrix(self, theta, axis_name):
        c = np.cos(theta * np.pi / 180.0)
        s = np.sin(theta * np.pi / 180.0)

        if axis_name == "x":
            return np.array([
                [1.0, 0.0, 0.0],
                [0.0, c, -s],
                [0.0, s, c]
            ], dtype=float)

        elif axis_name == "y":
            return np.array([
                [c, 0.0, s],
                [0.0, 1.0, 0.0],
                [-s, 0.0, c]
            ], dtype=float)

        elif axis_name == "z":
            return np.array([
                [c, -s, 0.0],
                [s, c, 0.0],
                [0.0, 0.0, 1.0]
            ], dtype=float)

        else:
            raise ValueError(f"Invalid axis_name: {axis_name}")

    def createCoordinateFrameMesh(self):
        unit = 30
        return (
            Arrow((0, 0, 0), (unit, 0, 0), c="red") +
            Arrow((0, 0, 0), (0, unit, 0), c="green") +
            Arrow((0, 0, 0), (0, 0, unit), c="blue") +
            Sphere((0, 0, 0), r=3, c="black")
        )

    def getLocalFrameMatrix(self, R, t):
        """
        Build a homogeneous transform:
        $$
        T = \begin{bmatrix} R & t \\ 0 & 1 \end{bmatrix}
        $$
        """
        R = np.asarray(R, dtype=float).reshape(3, 3)
        t = np.asarray(t, dtype=float).reshape(3, 1)

        T = np.eye(4, dtype=float)
        T[:3, :3] = R
        T[:3, 3:] = t
        return T

    def forward_kinematics(self, Phi):
        phi1, phi2, phi3, phi4 = Phi

        R_01 = self.RotationMatrix(phi1, "z")
        T_01 = self.getLocalFrameMatrix(R_01, self.arm_location)

        R_12 = self.RotationMatrix(phi2, "y")
        T_12 = self.getLocalFrameMatrix(R_12, np.array([[0.0], [0.0], [self.L1]], dtype=float))
        T_02 = T_01 @ T_12

        R_23 = self.RotationMatrix(phi3, "y")
        T_23 = self.getLocalFrameMatrix(R_23, np.array([[0.0], [0.0], [self.L2]], dtype=float))
        T_03 = T_02 @ T_23

        R_34 = self.RotationMatrix(phi4, "y")
        T_34 = self.getLocalFrameMatrix(R_34, np.array([[0.0], [0.0], [self.L3]], dtype=float))
        T_04 = T_03 @ T_34

        e = T_04[:3, 3]
        return T_01, T_02, T_03, T_04, e

    def get_pose_transforms(self, Phi):
        T_01, T_02, T_03, T_04, _ = self.forward_kinematics(Phi)

        T_00 = np.eye(4, dtype=float)
        T_00[:3, 3:] = self.arm_location

        return [T_00, T_01, T_02, T_03, T_04]

    def update_pose(self, Phi):
        transforms = self.get_pose_transforms(Phi)

        Base = self.source_parts[0].clone()
        Part1 = self.source_parts[1].clone()
        Part2 = self.source_parts[2].clone()
        Part3 = self.source_parts[3].clone()
        Part4 = self.createCoordinateFrameMesh()

        new_meshes = [Base, Part1, Part2, Part3, Part4]

        for mesh, T in zip(new_meshes, transforms):
            mesh.apply_transform(LinearTransform(T))

        self.meshes = new_meshes
        return self.meshes

    def jacobian_matrix(self, phi):
        phi = np.array(phi, dtype=float)
        _, _, _, _, e = self.forward_kinematics(phi)

        J_cols = []
        for i in range(4):
            phi_step = phi.copy()
            phi_step[i] += self.delta_phi
            _, _, _, _, e_step = self.forward_kinematics(phi_step)
            J_cols.append(((e_step - e) / self.delta_phi).reshape(3, 1))

        return np.hstack(J_cols)

    def ik_step(self, phi, step_scale=None):
        if step_scale is None:
            step_scale = self.target_lambda

        phi = np.array(phi, dtype=float)
        _, _, _, _, e = self.forward_kinematics(phi)

        J = self.jacobian_matrix(phi)
        J_pinv = np.linalg.pinv(J)

        delta_e = step_scale * (self.target - e)
        phi_new = phi + J_pinv @ delta_e

        _, _, _, _, e_new = self.forward_kinematics(phi_new)
        return phi_new, e_new

    def end_effector_xy(self, phi):
        _, _, _, _, e = self.forward_kinematics(phi)
        return e[:2]

    def set_target_from_xy(self, xy, z_height=80.0):
        self.target = np.array([float(xy[0]), float(xy[1]), float(z_height)], dtype=float)
```

# 9. Crowd Perception and Obstacle response

This cell includes:
- exit-based particle selection
- clustering
- motion prediction
- obstacle repulsion

The obstacle response is based on penetration into a circular safety region around the robot tip.

If a particle is inside the safety region, a repulsive velocity correction is applied in the outward radial direction.

For obstacle center $o$, particle position $p$, and safe radius $r_s$, the radial direction is:
$$
d = \frac{p - o}{\|p - o\|}
$$

and the repulsion strength increases as the particle moves deeper into the obstacle region.

```python
# 9. Crowd perception and obstacle response

import numpy as np

def obstacle_repulsion_from_circle(
    particle_xy,
    obstacle_center,
    obstacle_radius,
    gain=18.0,
    buffer=18.0,
    hard_core=6.0,
):
    """
    Compute a strong outward repulsive velocity correction from a circular obstacle.

    Parameters
    ----------
    particle_xy : (2,) array
    obstacle_center : (2,) array
    obstacle_radius : float
    gain : float
        Main repulsion gain.
    buffer : float
        Extra safety margin around the obstacle.
    hard_core : float
        Prevents division issues very close to the center.

    Returns
    -------
    repulsion : (2,) array
    """
    p = np.asarray(particle_xy, dtype=float)
    c = np.asarray(obstacle_center, dtype=float)

    dvec = p - c
    dist = np.linalg.norm(dvec)

    safe_radius = obstacle_radius + buffer

    if dist >= safe_radius:
        return np.zeros(2, dtype=float)

    if dist < 1e-8:
        # If exactly at the center, choose an arbitrary outward direction
        direction = np.array([1.0, 0.0], dtype=float)
        dist = 1e-8
    else:
        direction = dvec / dist

    # Penetration depth into the obstacle influence region
    penetration = safe_radius - dist

    # Stronger response near the core, smoother near the boundary
    strength = gain * (penetration / safe_radius) ** 2

    # Extra push if extremely close to the center
    if dist < hard_core:
        strength += gain * 1.5

    return strength * direction


def particles_near_exits(positions, exits, exit_radius=120.0):
    positions = np.asarray(positions, dtype=float)
    exits = [np.asarray(e, dtype=float) for e in exits]

    near = []
    for i, p in enumerate(positions):
        for ex in exits:
            if np.linalg.norm(p - ex) <= exit_radius:
                near.append(i)
                break

    return np.array(near, dtype=int)


def simple_kmeans(X, k=2, max_iter=25, seed=0):
    X = np.asarray(X, dtype=float)
    n = len(X)

    if n == 0:
        return None, None

    k = max(1, min(k, n))
    rng = np.random.default_rng(seed)
    centers = X[rng.choice(n, size=k, replace=False)].copy()
    labels = np.zeros(n, dtype=int)

    for _ in range(max_iter):
        dists = np.linalg.norm(X[:, None, :] - centers[None, :, :], axis=2)
        new_labels = np.argmin(dists, axis=1)

        new_centers = centers.copy()
        for j in range(k):
            pts = X[new_labels == j]
            if len(pts) > 0:
                new_centers[j] = pts.mean(axis=0)

        if np.array_equal(new_labels, labels) and np.allclose(new_centers, centers):
            break

        labels = new_labels
        centers = new_centers

    return labels, centers


def detect_exit_cluster(positions, velocities, exits, exit_radius=120.0, k=2, min_cluster_size=4):
    positions = np.asarray(positions, dtype=float)
    velocities = np.asarray(velocities, dtype=float)

    idx = particles_near_exits(positions, exits, exit_radius=exit_radius)

    if len(idx) < min_cluster_size:
        idx = np.arange(len(positions))

    X = positions[idx]
    V = velocities[idx]

    labels, centers = simple_kmeans(X, k=k, max_iter=25, seed=0)
    if labels is None:
        return None

    best = None
    best_score = -np.inf

    for j in range(len(centers)):
        mask = (labels == j)
        pts = X[mask]
        vel = V[mask]

        if len(pts) < min_cluster_size:
            continue

        centroid = pts.mean(axis=0)
        mean_vel = vel.mean(axis=0)
        speed = np.linalg.norm(mean_vel)

        score = len(pts) + 3.0 * speed

        if score > best_score:
            best_score = score
            best = {
                "centroid": centroid,
                "velocity": mean_vel,
                "size": len(pts),
                "indices": idx[mask],
            }

    return best


def predict_cluster_position(cluster, exits, horizon=10.0, exit_bias=0.2):
    if cluster is None:
        return None

    centroid = np.asarray(cluster["centroid"], dtype=float)
    velocity = np.asarray(cluster["velocity"], dtype=float)

    pred = centroid + horizon * velocity

    nearest_exit = min(
        exits,
        key=lambda ex: np.linalg.norm(centroid - np.asarray(ex, dtype=float))
    )
    nearest_exit = np.asarray(nearest_exit, dtype=float)

    pred = (1.0 - exit_bias) * pred + exit_bias * nearest_exit

    return pred
```

# 10. Crowd Simulation Setup

This cell defines the room, exits, particle initialization, and the crowd update rule. The robot tip acts as a circular obstacle in the 2D room. If a particle enters the obstacle influence region, a repulsive correction is added directly to its velocity.

```python
# 10. Crowd simulation setup

import numpy as np

# --------------------------------------------------
# Reproducibility
# --------------------------------------------------
np.random.seed(7)

# --------------------------------------------------
# Room geometry
# --------------------------------------------------
ROOM_W = 500.0
ROOM_H = 300.0

exits = [
    np.array([120.0, 0.0], dtype=float),
    np.array([380.0, 0.0], dtype=float),
]

# --------------------------------------------------
# Crowd initialization
# --------------------------------------------------
N = 80

positions = np.column_stack([
    np.random.uniform(80.0, 420.0, N),
    np.random.uniform(180.0, 280.0, N),
]).astype(float)

velocities = np.zeros((N, 2), dtype=float)

# --------------------------------------------------
# Simulation constants
# --------------------------------------------------
dt = 0.05
num_steps = 250

particle_speed = 18.0
particle_repulsion_radius = 12.0
particle_repulsion_gain = 4.0

robot_radius = 35.0
velocity_blend = 0.15

def nearest_exit(p):
    p = np.asarray(p, dtype=float)
    return min(exits, key=lambda ex: np.linalg.norm(p - ex))

def remove_escaped_particles(positions, velocities, exit_capture_radius=10.0):
    if len(positions) == 0:
        return positions, velocities

    keep_mask = np.ones(len(positions), dtype=bool)

    for i, p in enumerate(positions):
        for ex in exits:
            if np.linalg.norm(p - ex) <= exit_capture_radius:
                keep_mask[i] = False
                break

    return positions[keep_mask], velocities[keep_mask]

def update_crowd(positions, velocities, robot_xy=None):
    positions = np.asarray(positions, dtype=float).copy()
    velocities = np.asarray(velocities, dtype=float).copy()

    if len(positions) == 0:
        return positions, velocities

    positions_old = positions.copy()

    for i in range(len(positions)):
        p = positions_old[i]
        ex = nearest_exit(p)

        # Desired motion toward the nearest exit
        desired_dir = ex - p
        desired_norm = np.linalg.norm(desired_dir) + 1e-9
        desired_dir = desired_dir / desired_norm
        desired_vel = particle_speed * desired_dir

        # Smooth velocity update
        velocities[i] = (1.0 - velocity_blend) * velocities[i] + velocity_blend * desired_vel

        # Mild pairwise repulsion
        for j in range(len(positions_old)):
            if i == j:
                continue

            dvec = p - positions_old[j]
            dist = np.linalg.norm(dvec) + 1e-9

            if dist < particle_repulsion_radius:
                velocities[i] += particle_repulsion_gain * (dvec / dist)

        # Strong robot tip repulsion
        if robot_xy is not None:
            velocities[i] += obstacle_repulsion_from_circle(
                particle_xy=p,
                obstacle_center=robot_xy,
                obstacle_radius=robot_radius,
                gain=18.0,
                buffer=18.0,
                hard_core=6.0,
            )

    # Position update
    positions = positions + dt * velocities

    # Keep inside room
    positions[:, 0] = np.clip(positions[:, 0], 0.0, ROOM_W)
    positions[:, 1] = np.clip(positions[:, 1], 0.0, ROOM_H)

    # Remove escaped particles
    positions, velocities = remove_escaped_particles(
        positions,
        velocities,
        exit_capture_radius=10.0
    )

    return positions, velocities

print("Crowd setup complete.")
print(f"Initial particle count: {len(positions)}")
print(f"Room size: {ROOM_W} x {ROOM_H}")
print(f"Number of exits: {len(exits)}")
```

# 11. Closed-loop Robot–crowd Simulation

This cell couples the crowd simulation and robot control into a closed-loop system.

At each timestep:
1. The current robot end-effector position is projected into the 2D room.
2. That projected tip position acts as the crowd-repelling obstacle.
3. The crowd state is updated using this obstacle.
4. A cluster near the exits is detected.
5. The future position of the cluster is predicted.
6. The robot updates its target and moves using inverse kinematics.

The crowd-repelling obstacle is centered at the robot tip:
$$
o^{(t)} = \operatorname{proj}_{xy}(e^{(t)})
$$

and particles are repelled from a circular region of radius $r_{\text{robot}}$ around that point.

This means the robot tip is the active part of the robot that influences the crowd.

```python
# 11. Closed-loop robot–crowd simulation

import matplotlib.pyplot as plt
import os
import glob
import numpy as np

frames_dir = "crowd_robot_frames"
os.makedirs(frames_dir, exist_ok=True)

for f in glob.glob(os.path.join(frames_dir, "*.png")):
    os.remove(f)

# --------------------------------------------------
# Robot inside the room
# --------------------------------------------------
crowd_arm_location = np.array([[ROOM_W * 0.5], [ROOM_H * 0.75], [BaseH]])
crowd_robot = RobotArm(L, parts, crowd_arm_location)

phi_crowd = np.array([10.0, -20.0, 45.0, 0.0], dtype=float)

# --------------------------------------------------
# Histories
# --------------------------------------------------
phi_history_crowd = []
position_history = []
cluster_history = []
pred_history = []
robot_xy_history = []
robot_tip_xyz_history = []
robot_obstacle_radius_history = []

# --------------------------------------------------
# Simulation loop
# --------------------------------------------------
for t in range(num_steps):

    # 1) Current end-effector position
    _, _, _, _, e_now = crowd_robot.forward_kinematics(phi_crowd)
    robot_tip_xyz_now = np.array(e_now, dtype=float)

    # Tip projected into the 2D room
    robot_xy_now = robot_tip_xyz_now[:2].copy()
    robot_xy_now = np.clip(robot_xy_now, [0.0, 0.0], [ROOM_W, ROOM_H])

    # 2) Crowd update with tip as obstacle
    positions, velocities = update_crowd(
        positions,
        velocities,
        robot_xy=robot_xy_now
    )

    # 3) Detect cluster near exits
    cluster = detect_exit_cluster(
        positions,
        velocities,
        exits,
        exit_radius=120.0,
        k=2,
        min_cluster_size=4,
    )

    pred_xy = None
    cluster_indices = np.array([], dtype=int)

    if cluster is not None:
        cluster_indices = np.asarray(cluster["indices"], dtype=int)

        # 4) Predict future cluster position
        pred_xy = predict_cluster_position(
            cluster,
            exits,
            horizon=10.0,
            exit_bias=0.2,
        )
        pred_xy = np.clip(pred_xy, [0.0, 0.0], [ROOM_W, ROOM_H])

        # 5) Set IK target
        crowd_robot.set_target_from_xy(pred_xy, z_height=80.0)

    # 6) Reactive IK updates
    for _ in range(8):
        phi_crowd, _ = crowd_robot.ik_step(phi_crowd, step_scale=0.03)

    # Updated tip after IK
    _, _, _, _, e_new = crowd_robot.forward_kinematics(phi_crowd)
    robot_tip_xyz_new = np.array(e_new, dtype=float)

    robot_xy_new = robot_tip_xyz_new[:2].copy()
    robot_xy_new = np.clip(robot_xy_new, [0.0, 0.0], [ROOM_W, ROOM_H])

    # --------------------------------------------------
    # Record history
    # --------------------------------------------------
    phi_history_crowd.append(phi_crowd.copy())
    position_history.append(positions.copy())
    cluster_history.append(cluster)
    pred_history.append(pred_xy.copy() if pred_xy is not None else None)
    robot_xy_history.append(robot_xy_new.copy())
    robot_tip_xyz_history.append(robot_tip_xyz_new.copy())
    robot_obstacle_radius_history.append(robot_radius)

    # --------------------------------------------------
    # 2D frame
    # --------------------------------------------------
    fig, ax = plt.subplots(figsize=(8, 5))

    if len(positions) > 0:
        ax.scatter(positions[:, 0], positions[:, 1], s=16, label="particles")

    for ex in exits:
        ax.scatter(ex[0], ex[1], marker="s", s=120, label="exit")

    if len(cluster_indices) > 0:
        pts = positions[cluster_indices]
        ax.scatter(pts[:, 0], pts[:, 1], s=24, color="orange", label="cluster")

    if cluster is not None:
        c = cluster["centroid"]
        ax.scatter(c[0], c[1], marker="x", s=140, color="black", label="centroid")

    if pred_xy is not None:
        ax.scatter(pred_xy[0], pred_xy[1], marker="*", s=180, color="green", label="prediction")

    # Repulsion circle centered at robot tip projection
    circ = plt.Circle(
        (robot_xy_new[0], robot_xy_new[1]),
        robot_radius,
        fill=False,
        color="red",
        linewidth=2
    )
    ax.add_patch(circ)

    ax.scatter(
        robot_xy_new[0],
        robot_xy_new[1],
        color="red",
        s=80,
        label="robot tip obstacle"
    )

    ax.set_xlim(0, ROOM_W)
    ax.set_ylim(0, ROOM_H)
    ax.set_aspect("equal")
    ax.set_title(f"Step {t} | Particles: {len(positions)}")

    handles, labels = ax.get_legend_handles_labels()
    unique = dict(zip(labels, handles))
    ax.legend(unique.values(), unique.keys(), loc="upper right")

    plt.tight_layout()
    plt.savefig(os.path.join(frames_dir, f"frame_{t:03d}.png"), dpi=120)
    plt.close(fig)

print("Simulation complete.")
print(f"Remaining particles: {len(positions)}")
```

# 14. Render 3D robot motion frames

This cell renders a 3D reconstruction of the closed-loop simulation using `matplotlib`.

For each saved simulation state, it displays:
- the robot arm as a kinematic chain
- crowd particles on the floor plane
- exit markers
- the predicted interception target
- the crowd-repelling tip obstacle as a 3D sphere centered at the end-effector

The repelling region is visualized as a sphere of radius $r_{\text{robot}}$ centered at the robot tip:
$$
\|x - e^{(t)}\| = r_{\text{robot}}
$$

This makes the active interaction region of the robot visible in the 3D render.

```python
# 14. Render 3D robot motion frames

import os
import glob
import numpy as np
import matplotlib.pyplot as plt

required_sequences = [
    "phi_history_crowd",
    "position_history",
    "pred_history",
    "robot_tip_xyz_history",
    "robot_obstacle_radius_history",
]
for name in required_sequences:
    if name not in globals() or globals()[name] is None or len(globals()[name]) == 0:
        raise ValueError(f"{name} is empty or not defined. Run the closed-loop simulation cell first.")

if "crowd_robot" not in globals() or crowd_robot is None:
    raise ValueError("crowd_robot is not defined. Run the closed-loop simulation cell first.")

frames_3d_dir = "robot_3d_frames"
os.makedirs(frames_3d_dir, exist_ok=True)

for f in glob.glob(os.path.join(frames_3d_dir, "*.png")):
    os.remove(f)

def extract_joint_positions(robot, phi):
    T_01, T_02, T_03, T_04, e = robot.forward_kinematics(phi)

    base = robot.arm_location.reshape(3)
    p1 = T_01[:3, 3]
    p2 = T_02[:3, 3]
    p3 = T_03[:3, 3]
    p4 = T_04[:3, 3]

    return np.vstack([base, p1, p2, p3, p4])

def draw_sphere_wireframe(ax, center, radius, color="red", alpha=0.12, wire_alpha=0.35):
    u = np.linspace(0, 2 * np.pi, 30)
    v = np.linspace(0, np.pi, 20)

    x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
    y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
    z = center[2] + radius * np.outer(np.ones_like(u), np.cos(v))

    ax.plot_surface(x, y, z, color=color, alpha=alpha, linewidth=0, shade=False)
    ax.plot_wireframe(x, y, z, color=color, alpha=wire_alpha, linewidth=0.4)

frame_stride = 1
render_indices = list(range(0, len(phi_history_crowd), frame_stride))

for out_idx, sim_idx in enumerate(render_indices):
    phi = np.array(phi_history_crowd[sim_idx], dtype=float)
    crowd_xy = position_history[sim_idx]
    pred_xy = pred_history[sim_idx]
    tip_xyz = np.array(robot_tip_xyz_history[sim_idx], dtype=float)
    obstacle_radius = float(robot_obstacle_radius_history[sim_idx])

    joint_pts = extract_joint_positions(crowd_robot, phi)

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection="3d")

    # Crowd particles on floor plane
    if len(crowd_xy) > 0:
        ax.scatter(
            crowd_xy[:, 0],
            crowd_xy[:, 1],
            np.full(len(crowd_xy), 2.0),
            s=8,
            c="black",
            alpha=0.35
        )

    # Exits
    for ex in exits:
        ax.scatter(ex[0], ex[1], 3.0, s=70, c="green", marker="s")

    # Predicted interception target
    if pred_xy is not None:
        ax.scatter(pred_xy[0], pred_xy[1], 12.0, s=120, c="gold", marker="*")

    # Robot skeleton
    ax.plot(
        joint_pts[:, 0],
        joint_pts[:, 1],
        joint_pts[:, 2],
        linewidth=3,
        marker="o",
        markersize=6
    )

    # End-effector
    ax.scatter(
        tip_xyz[0],
        tip_xyz[1],
        tip_xyz[2],
        s=90,
        c="red"
    )

    # Repulsion sphere centered at the tip
    draw_sphere_wireframe(ax, tip_xyz, obstacle_radius, color="red", alpha=0.10, wire_alpha=0.22)

    # Floor outline
    floor_x = [0, ROOM_W, ROOM_W, 0, 0]
    floor_y = [0, 0, ROOM_H, ROOM_H, 0]
    floor_z = [0, 0, 0, 0, 0]
    ax.plot(floor_x, floor_y, floor_z, linewidth=1.5, alpha=0.6)

    ax.set_xlim(0, ROOM_W)
    ax.set_ylim(0, ROOM_H)
    ax.set_zlim(0, 450)

    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")

    ax.view_init(elev=28, azim=-60)
    ax.set_title(f"3D Robot Motion - step {sim_idx}")

    plt.tight_layout()
    out_path = os.path.join(frames_3d_dir, f"frame_{out_idx:03d}.png")
    plt.savefig(out_path, dpi=140)
    plt.close(fig)

actual_files = sorted(glob.glob(os.path.join(frames_3d_dir, "frame_*.png")))
print(f"Actually created {len(actual_files)} 3D frames in: {frames_3d_dir}")
```
